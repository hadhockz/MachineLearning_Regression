```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the dataset
dts = pd.read_csv("insurance_pre.csv")
```

```python
dts
```

```python
dts = pd.get_dummies(dts,drop_first=True)
```

```python
dts
```

```python
#columns rename-ing
dts.rename(columns={"sex_male":"sex","smoker_yes":"smoker","charges":"premium_amt"},inplace=True)
```

```python
dts
```

```python
dts.columns
```

```python
features = dts[["age","bmi","children","sex","smoker"]]
target = dts[["premium_amt"]]
```

```python
features
```

```python
target
```

```python
#import RF & GRIDSEARCH Libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
hyperparam_grid = {"criterion":["absolute_error", "friedman_mse", "poisson"],
                   "max_features" : ["sqrt","log2"],
                   "n_estimators" : [10,100]}

grid = GridSearchCV(RandomForestRegressor(), hyperparam_grid, 
                    refit = True, verbose = 3, n_jobs = -1)

#model fitting for grid search
grid.fit(features,target)
```

```python
#printing best parameter after iterations from mentioned hyperparameters.

re = grid.cv_results_
print("The R2 score value for best parameter {}:".format(grid.best_params_))
```

```python
#Table building for data's
table = pd.DataFrame.from_dict(re)
#os library to perform file management process.
import os
#Mentioning the directory to save the dataframe in a formatted file
directory = '/home/had/jnotebook/HOPE_py_training/MachineLearning-Regression/GridSearchCV'
#saving .xlsx file in above mentioned directory
file_path = os.path.join(directory,"iterations_RF.xlsx")
#converting dataframe to .xlsx format.
table.to_excel(file_path, index=False)
```
