```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the dataset
dts = pd.read_csv("insurance_pre.csv")
```

```python
dts
```

```python
dts = pd.get_dummies(dts,drop_first=True)
```

```python
dts
```

```python
#columns rename-ing
dts.rename(columns={"sex_male":"sex","smoker_yes":"smoker","charges":"premium_amt"},inplace=True)
```

```python
dts
```

```python
dts.columns
```

```python
features = dts[["age","bmi","children","sex","smoker"]]
target = dts[["premium_amt"]]
```

```python
features
```

```python
target
```

```python
#import RF & GRIDSEARCH Libraries
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
hyperparam_grid = {"kernel":["rbf","sigmoid","linear"],
                   "C" : [10,50,100],
                   "gamma" : ["auto","scale"]}

grid = GridSearchCV(SVR(), hyperparam_grid,cv = 3, 
                    refit = True, verbose = 3, n_jobs = -1)

#model fitting for grid search
grid.fit(features,target)
```

```python
#printing best parameter after iterations from mentioned hyperparameters.

re = grid.cv_results_
print("The R2 score value for best parameter {}:".format(grid.best_params_))
```

```python
#Table building for data's
table = pd.DataFrame.from_dict(re)
#os library to perform file management process.
import os
#Directory to save the xlsx file
directory = '/home/had/jnotebook/HOPE_py_training/ML-Linear Regression-codes/GridSearchCV'
#file path for the xlsx file
file_path = os.path.join(directory,"iterations_SVR.xlsx")
#Save the dataframe to xlsx
table.to_excel(file_path, index=False)
```
