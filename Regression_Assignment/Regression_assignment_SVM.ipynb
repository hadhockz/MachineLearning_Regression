```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#Storing the .CSV information into a variable
dts = pd.read_csv("insurance_pre.csv")
```

```python
dts
```

```python
#viewing the total no of columns
dts.columns
```

```python
dts = pd.get_dummies(dts)
```

```python
dts
```

```python
feature = dts[["age","bmi","children","sex_female","sex_male","smoker_yes","smoker_no"]]
feature
```

```python
target = dts[["charges"]]
target
```

```python
#splitting the train & test dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(feature,target,test_size=0.33,random_state=0)
```

```python
#Printing training dataset
X_train, Y_train
```

```python
#Printing testing dataset.
Y_train, Y_test
```

```python
X_train
```

```python
#Standardisation procedure to maintain least different between inputs.
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
#Model creation and training
from sklearn.svm import SVR
model_create = SVR(kernel = "sigmoid", C=10)
model_create.fit(X_train, Y_train)
```

```python
#checking the values of slope 'm' and bias 'c' values
#'coef_' can be used when kernel = "linear" 
#otherwise showing AttributeError: coef_ is only available when using a linear kernel
#'intercept_' should be used when kernel = "rbf" (radial bias/Intercept factor)
bias =  model_create.intercept_
bias
```

```python
slope = model_create.n_support_
slope
```

```python
#support vector datapoints count.
model_create.support_
```

```python
Y_pred = model_create.predict(X_test)
```

```python
#Evaluation metric phase
from sklearn.metrics import r2_score
rscore= r2_score(Y_test,Y_pred)
```
