```python
#importing libraries
import pandas as pd
import numpy as np
import pickle
```

```python
# Reading dataset
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
#Removing the presence of dummies in state column & converting categorical value to numerical.
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#Column renaming
dataset.rename(columns={"State_Florida":"Florida","State_New York":"NewYork"},inplace=True)
```

```python
#printing the dataset columns
dataset.columns
```

```python
#separating the input and output values
feature = dataset[["R&D Spend","Administration","Marketing Spend","Florida","NewYork"]]
target = dataset[["Profit"]]
```

```python
feature,target
```

```python
#saving dataset
np.savetxt('dataset_file.txt', dataset)
np.save('dataset_file.npy',dataset)
```

```python
#splitting the train and test set values
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(feature,target,test_size = 0.33, random_state = 0)
```

```python
X_train
```

```python
target
```

```python
#saving the train and test data.
with open("train_data.pkl", "wb") as file:
    pickle.dump((X_train,Y_train), file)
with open("test_data.pkl", "wb") as file:
    pickle.dump((X_test,Y_test), file)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from joblib import dump
dump(sc,"standarad_scaler.joblib")
```

```python
X_train
```

```python
X_test
```

```python
#Model creation & training
from sklearn.svm import SVR
model_svm = SVR(kernel = "rbf",C=1000)
model_svm.fit(X_train,Y_train)
```

```python
model_svm.intercept_
```

```python
model_svm.n_support_
```

```python
#Prediction using preprocessed input dataset
Y_pred = model_svm.predict(X_test)
```

```python
#Prediction in Raw dataset
print("Profit are =\n", Y_pred)
```

```python
Y_test
```

```python
#Evaluation metrics using Raw test_input & Raw_output
from sklearn.metrics import r2_score
r2value = r2_score(Y_test,Y_pred)
```

```python
#Evaluation metrics result
r2value
```

```python
#Saving best model phase in .sav file format using pickle library
import pickle
filename = "bestmodel_svm_preinput.sav"
pickle.dump(model_svm,open(filename,"wb"))
```
