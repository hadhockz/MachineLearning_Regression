```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Dataset Reading
dts = pd.read_csv("50_Startups.csv")
```

```python
dts
```

```python
#Viewing columns in table
dts.columns
```

```python
#converting the categorical value to numerical and removal of redundancy
dts = pd.get_dummies(dts,drop_first=True)
```

```python
dts
```

```python
#separating input and output
features = dts[["R&D Spend","Administration","Marketing Spend","State_Florida","State_New York"]]
target = dts[["Profit"]]
```

```python
features
```

```python
target
```

```python
#splitting train and test set's
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(features,target,test_size=0.33,random_state=0)
```

```python
X_train, X_test
```

```python
Y_train, Y_test
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion = "mae", splitter ="best", max_features ="sqrt")
regressor = regressor.fit(X_train,Y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
```

```python
r_score
```
